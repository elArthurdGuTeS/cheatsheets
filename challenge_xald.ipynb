#PRIMERO

```python
#Importar librerias usadas
import requests
import pandas as pd
```

```python
#Llamando a la API con la libreria requests
url = 'https://api.stackexchange.com/2.2/search?order=desc&sort=activity&intitle=perl&site=stackoverflow'
response = requests.get(url).json()
```

```python
#Ver el contenido de la API en formato JSON
response['items']
```

```python
#Creando el cuadro de datos 
df = pd.DataFrame(columns=["name","answer_count","views","old_answer","new_answer","reputation"])
```

```python

#Buscando la informacion de cada persona mediante un loop
for person in response['items']:
  if person['owner']['user_type'] == 'registered':
    name = person['owner']['display_name']
    answer_count = person['answer_count']
    views = person['view_count']
    old_answer = person['creation_date']
    new_answer = person['last_activity_date']
    reputation = person['owner']['reputation']
    #imprimir resultados previos
    #print(answered)
    #print(views)
    #print(old_answer)
    #print(new_answer)
    #print(reputation)

    #Guardando los datos con la libreria pandas
    df = df.append({'name': name, 'answer_count': answer_count, 'views': views, 'old_answer': old_answer, 'new_answer': new_answer, 'reputation': reputation}, ignore_index=True)
```

```python
#imprimir resultados previos
df
```

```python
#Buscando personas con respuestas contestadas
answered = df[df['answer_count'] > 4]
answered
```

```python
#Buscando personas con respuestas no contestadas
answered = df[df['answer_count'] == 0]
answered
```

```python
#Buscando respuestas con menos vistas
views = df[df['views'] < 30]
views
```

```python
#Buscando la respuesta mas vieja
old_answer = df[df['old_answer'] < 1270000000]
old_answer
```

```python
#Buscando la respuesta mas nueva
new_answer = df[df['new_answer'] > 1660950000]
new_answer
```

```python
#Buscando la respuesta con mas reputacion
reputation = df[df['reputation'] > 70000]
reputation
```
